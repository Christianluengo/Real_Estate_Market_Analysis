```python
import pandas as pd
import sqlalchemy as alch
from sqlalchemy.exc import SQLAlchemyError
from getpass import getpass # sistema seguro para gestión de contraseñas
```

```python

nombre_bbdd = 'prueba4'
contraseña = 'admin'

def create_db():

    engine = alch.create_engine(f"mysql+pymysql://root:{contraseña}@localhost")
    try:
        engine.execute(f"CREATE DATABASE IF NOT EXISTS {nombre_bbdd};")
        print(f"---------- '{nombre_bbdd.upper()}' ha sido creada con éxito ----------")
    except:
        print("ERROR:  La BBDD ya existe")



def conexion_db():
    return alch.create_engine(f"mysql+pymysql://root:{contraseña}@localhost/{nombre_bbdd}")
```

```python

table_region = '''
CREATE TABLE IF NOT EXISTS `estate_db`.`Region` (
  `idregion` INT NOT NULL,
  `region` VARCHAR(45) NOT NULL,
  PRIMARY KEY (`idregion`));
'''

table_province = """
CREATE TABLE IF NOT EXISTS `estate_db`.`Provincia` (
  `idProvincia` INT NOT NULL,
  `province` VARCHAR(45) NOT NULL,
  PRIMARY KEY (`idProvincia`));
"""

table_city = """
CREATE TABLE IF NOT EXISTS `estate_db`.`Municipio` (
  `idMunicipio` INT NOT NULL,
  `city` VARCHAR(45) NOT NULL,
  PRIMARY KEY (`idMunicipio`));
"""

table_zipcode = """
CREATE TABLE IF NOT EXISTS `estate_db`.`Codigo_postal` (
  `idzipcode` INT NOT NULL,
  `zipcode` VARCHAR(5) NOT NULL,
  PRIMARY KEY (`idzipcode`));
"""

table_coordinates = """
CREATE TABLE IF NOT EXISTS `estate_db`.`Coordenadas` (
  `idcordenadas` INT NOT NULL AUTO_INCREMENT,
  `latitude` FLOAT NOT NULL,
  `longitude` FLOAT NOT NULL,
  `Region_idregion` INT NOT NULL,
  `Provincia_idProvincia` INT NOT NULL,
  `Municipio_idMunicipio` INT NOT NULL,
  `Codigo_postal_idzipcode` INT NOT NULL,
  PRIMARY KEY (`idcordenadas`),
  INDEX `fk_Coordenadas_Region1_idx` (`Region_idregion` ASC) VISIBLE,
  INDEX `fk_Coordenadas_Provincia1_idx` (`Provincia_idProvincia` ASC) VISIBLE,
  INDEX `fk_Coordenadas_Municipio1_idx` (`Municipio_idMunicipio` ASC) VISIBLE,
  INDEX `fk_Coordenadas_Codigo_postal1_idx` (`Codigo_postal_idzipcode` ASC) VISIBLE,
  CONSTRAINT `fk_Coordenadas_Region1`
    FOREIGN KEY (`Region_idregion`)
    REFERENCES `estate_db`.`Region` (`idregion`)
    ON DELETE NO ACTION
    ON UPDATE NO ACTION,
  CONSTRAINT `fk_Coordenadas_Provincia1`
    FOREIGN KEY (`Provincia_idProvincia`)
    REFERENCES `estate_db`.`Provincia` (`idProvincia`)
    ON DELETE NO ACTION
    ON UPDATE NO ACTION,
  CONSTRAINT `fk_Coordenadas_Municipio1`
    FOREIGN KEY (`Municipio_idMunicipio`)
    REFERENCES `estate_db`.`Municipio` (`idMunicipio`)
    ON DELETE NO ACTION
    ON UPDATE NO ACTION,
  CONSTRAINT `fk_Coordenadas_Codigo_postal1`
    FOREIGN KEY (`Codigo_postal_idzipcode`)
    REFERENCES `estate_db`.`Codigo_postal` (`idzipcode`)
    ON DELETE NO ACTION
    ON UPDATE NO ACTION);
"""

table_subtype = """
CREATE TABLE IF NOT EXISTS `estate_db`.`Tipo_vivienda` (
  `idTipo_vivienda` INT NOT NULL,
  `subtypeId` INT NOT NULL,
  PRIMARY KEY (`idTipo_vivienda`));
"""

table_estate = """
CREATE TABLE IF NOT EXISTS `estate_db`.`Vivienda` (
  `idtable1` INT NOT NULL AUTO_INCREMENT,
  `id` VARCHAR(45) NOT NULL,
  `date` DATE NOT NULL,
  `advertiser` VARCHAR(45) NULL,
  `descriptions` VARCHAR(355) NULL,
  `rooms` INT NOT NULL,
  `bathrooms` INT NOT NULL,
  `surface` INT NOT NULL,
  `energy_certificate` INT(1) NOT NULL,
  `antiquity` INT NOT NULL,
  `floor` INT NOT NULL,
  `surfaceland` INT NOT NULL,
  `price` INT NOT NULL,
  `Tipo_vivienda_idTipo_vivienda` INT NOT NULL,
  `Coordenadas_idcordenadas` INT NOT NULL,
  PRIMARY KEY (`idtable1`, `Tipo_vivienda_idTipo_vivienda`),
  INDEX `fk_Vivienda_Tipo_vivienda_idx` (`Tipo_vivienda_idTipo_vivienda` ASC) VISIBLE,
  INDEX `fk_Vivienda_Coordenadas1_idx` (`Coordenadas_idcordenadas` ASC) VISIBLE,
  CONSTRAINT `fk_Vivienda_Tipo_vivienda`
    FOREIGN KEY (`Tipo_vivienda_idTipo_vivienda`)
    REFERENCES `estate_db`.`Tipo_vivienda` (`idTipo_vivienda`)
    ON DELETE NO ACTION
    ON UPDATE NO ACTION,
  CONSTRAINT `fk_Vivienda_Coordenadas1`
    FOREIGN KEY (`Coordenadas_idcordenadas`)
    REFERENCES `estate_db`.`Coordenadas` (`idcordenadas`)
    ON DELETE NO ACTION
    ON UPDATE NO ACTION)
ENGINE = InnoDB;
"""

list_tables = ['table_region', 'table_province', 'table_city', 'table_zipcode', 'table_coordinates', 'table_subtype', 'table_estate']
```

```python
def crear_insertar_tabla(query):
    engine = conexion_db()


    try:
        engine.execute(query)
        
    except SQLAlchemyError as e:
        error = str(e.__dict__['orig'])
        return error
```
